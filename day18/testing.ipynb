```python
import matplotlib.pyplot as plt
import shapely

from day18 import DigPlan

def plot_polygon(poly, ax):
    # Plot exterior
    x, y = poly.exterior.xy
    ax.fill(x, y, color='lightblue', edgecolor='blue')

    # Plot interior (holes)
    for interior_ring in poly.interiors:
        x, y = interior_ring.xy
        ax.fill(x, y, color='white', edgecolor='blue') # Fill holes with white

def plot_polygons(polys):
    fig, ax = plt.subplots(figsize=(10, 10))
    ax.set_aspect('equal')

    for p in polys:
        plot_polygon(p, ax)
    plt.show()
```

```python
plan = DigPlan.from_file("test.txt")
plot_polygons([plan.trench])
```
