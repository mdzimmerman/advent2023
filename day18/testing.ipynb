```python
import matplotlib.pyplot as plt
import shapely

from day18 import DigPlan
```

```python
plan = DigPlan.from_file("test.txt")
print(plan.trench)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.set_aspect('equal')

def plot_polygon(poly, ax):
    # Plot exterior
    x, y = poly.exterior.xy
    ax.fill(x, y, color='lightblue', edgecolor='blue')

    # Plot interior (holes)
    for interior_ring in poly.interiors:
        x, y = interior_ring.xy
        ax.fill(x, y, color='white', edgecolor='blue') # Fill holes with white

#plot_polygon(plan.trench, ax)
#for p in plan.segments:
#    plot_polygon(p, ax)
trench = shapely.union_all(plan.segments)
plot_polygon(trench, ax)
plt.show()
```
